**Calculate the inverse of a matrix and LU decomposition**

1. Use numpy's $\tt linalg$ package to calculate the inverse of the ff. matrix *(5 pts.)*: 

$ \begin{equation} 
{\bf A} = 
\begin{pmatrix} 
4 & -2 & 1 \\
3 & 6 & -4 \\
2 & 1 & 8 
\end{pmatrix} \end{equation} $

2. Check your inverse in both directions; that is, check that ${\bf AA}^{−1} = {\bf A}^{−1}{\bf A} = {\bf I}$. Up to how many number of decimal places is this true? *(5 pts.)*

3. Let this matrix $\bf A$ be used to describe three simultaneous linear equations, ${\bf Ax = b}$. Solve for the vector $\bf x$ given
${\bf b} = \begin{pmatrix} 
12 \\ 
-25 \\ 
32 
\end{pmatrix}$ using: (i) the numerical inverse calculated in (1) and (ii) LU decomposition and backsubstitution with numpy's $\tt linalg$ package. *(5 pts. each)*


**Solution**

Since we are using numpy's $\tt linalg$ package, we first have to input numpy itself. Then take the given 3 x 3 matrix, make an array, and compute for the inverse.

```python
# Importing required package
import numpy as np

# Storing the given matrix in an array
A = np.array([[4, -2, 1],
              [3, 6, -4],
              [2, 1, 8]], float)

# Calculating the inverse yung the package
A_inv = np.linalg.inv(A)
print(A_inv)
```

To check if ${\bf AA}^{−1} = {\bf A}^{−1}{\bf A} = {\bf I}$, we use numpy's $\tt dot$ method.

```python
# For AA^-1
print(np.matmul(A,A_inv), "\n")

# For A^-1A
print(np.matmul(A_inv, A))
```

An identity matrix is where the principal diagonal are ones and the rest of the elements are zeros. From the results, we can just approximate the other values as zero. Based on the otthe values, this is true up to the 161t, 17th, and 18th places.

Next is we use matrix $A$ to describe three simultaneous linear equations. Solving for $x$ numerically, where in $Ax=b$, we can just multiply the inverse of A on both sides. That will be as follows:

```python
# Numerical inverse calculation
# Ax = b 
# A^-1 Ax = A^-1 b
# Ix = A^-1 b
b = np.array([[12],
              [-25],
              [32]], float)
x = np.matmul(A_inv, b)
print(x)
```

To check the values of $x$, let's now try the other method: the LU decomposition and backsubstition.

The matrices and vectors are already stored as arrays, hence we can move on to the elimination/decomposition right away. It it where the each row of the matrix is normalized, then a backsubstitution is applied to get the exact values of $x$. Or we can just use the $\tt linalg$ function directly from numpy.

```python
from numpy.linalg import solve
x = solve(A,b)
print(x)
```

### PROBLEM 2 (15 pts.; 50/50 for code & discussion)

**Solve for eigenvalues and eigenvectors of a symmetric matrix**

1. Use numpy's $\tt linalg$ package to find the eigenvalues and eigenvectors of the ff. symmetrix matrix *(10 pts.)*: 

$ \begin{equation} 
{\bf A} = 
\begin{pmatrix} 
8 & -2 & -20 \\
-2 & -3 & -9 \\
-20 & -9 & -3 
\end{pmatrix} \end{equation} $

2. Check your answer numerically; that is, check that $\bf AV = VD$, where the $N$ x $N$ matrix $\bf V$ has the eigenvectors as its columns and $\bf D$ is the diagonal matrix with the eigenvalues $\lambda_i$ as its diagonal entries. *(5 pts.)*

**Solution**

First is to import the packages needed from $\tt numpy$. And in order to solve for the matrix, we have store its values into a three-dimensionsal array. Then, use the $\tt eigh()$ method to calculate for the eigenvalues and eigenvectors of the matrix.

```python
# Importing necessary libraries
from numpy import array
from numpy.linalg import eigh 

# Storing the matrix in an array
A = array([[ 8, -2, -20 ], 
            [ -2, -3, -9 ], 
            [ -20, -9, -3]], float) 

# Calling the eigh() method to calculate both the eigenvalues and eigenvectors
x,V = eigh(A)

print('Eigenvalues: ', x, "\n") # Printing the eigenvalues
print('Eigenvectors: ', V) # Printing the eigenvectors
```

The code results to two arrays: the one-dimensional array containing the eigenvalues which are $-21.99606195$, $0.16021522$, and $23.83584674$ , and the three-dimensional array whose columns are the corresponding eigenvectors.

To check our answer numerically, we show that $\bf AV = VD$. $V$ and $A$ are already defined, hence we just need to construct the diagonal matrix D.

```python
# Left-hand side A V
print('AV: ', np.matmul(A,V), "\n")

# Constructing the diagonal matrix
D = array([[ -21.99606195, 0, 0 ], 
            [ 0, 0.16021522, 0 ], 
            [ 0, 0, 23.83584674]], float) 

# Right-hand side V D
print('VD: ', np.matmul(V,D))
```

As shown, the values from the left-hand side and right-hand side which means that $\bf AV = VD$.

**Physics Application: Rigid body rotations with matrices**

Consider the cube in this figure:

fig_cube.png

The cube has an inertia tensor for axes at the center:

$ \begin{equation} 
{\bf I} = 
\begin{pmatrix} 
2/3 & -1/4 & -1/4 \\
-1/4 & 2/3 & -1/4 \\
-1/4 & -1/4 & 2/3 
\end{pmatrix}. \end{equation} $

The cube is rotated with the three angular velocities:

$ \begin{equation} 
{\bf \omega} = 
\begin{pmatrix}
1 \\ 0 \\ 0
\end{pmatrix}, \ {\bf \omega} = 
\begin{pmatrix}
0 \\ 1 \\ 0
\end{pmatrix} \ {\bf \omega} = 
\begin{pmatrix}
1 \\ 1 \\ 1
\end{pmatrix}
\end{equation} $

1. Compute the angular momentum vector $\bf L$ via the matrix multiplication ${\bf L} = {\bf I\omega}$ for each case. (5 pts.)

2. Make three 3-D plots for ${\bf \omega}$ and ${\bf L}$ for each case. Label and describe your plots clearly. (10 pts.)

**Solution**

We first store the matrix $I$ and vectors into arrays. Then we solve for the  angular momentum vector $L = I\omega$ of the three angular velocities.

```python
# Storing the I matrix in an array
I = array([[ 2/3, -1/4, -1/4 ], 
            [ -1/4, 2/3, -1/4 ], 
            [ -1/4, -1/4, 2/3]], float) 

# Storing the angular velocities in arrays
w1 = array([[ 1 ], 
            [ 0 ], 
            [ 0]], float) 
w2 = array([[ 0 ], 
            [ 1 ], 
            [ 0]], float)
w3 = array([[ 1 ], 
            [ 1 ], 
            [ 1 ]], float)

# Solving for the angular momentum vector
L1 = np.matmul(I,w1)
L2 = np.matmul(I,w2)
L3 = np.matmul(I,w3)

# Printing the results for each
print("L1: ", L1, "\n")
print("L2: ", L2, "\n")
print("L3: ", L3, "\n")
```

Next is we make a 3D plot for ${\bf \omega}$ and ${\bf L}$ of each case. First, we import the neccessary libraries.

```python
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
```

Then we use the $\tt ax.quiver()$ method to make the 3D plots.

```python
# Plotting the first angular velocity w1 and angular momentum L1
ax = plt.figure().add_subplot(projection='3d')
ax.quiver(0, 0, 0, L1[0,0], L1[1,0], L1[2,0], color = 'steelblue', length = 0.1, label = 'L1')
ax.quiver(0, 0, 0, w1[0,0], w1[1,0], w1[2,0], color = 'brown', length = 0.1, label = 'w1')
ax.legend()
```

```python
# Plotting the second angular velocity w2 and angular momentum L2
ax = plt.figure().add_subplot(projection='3d')
ax.quiver(0, 0, 0, L2[0,0], L2[1,0], L2[2,0], color = 'steelblue', length = 0.1, label = 'L2')
ax.quiver(0, 0, 0, w2[0,0], w2[1,0], w2[2,0], color = 'brown', length = 0.1, label = 'w2')
ax.legend()
```

```python
# Plotting the third angular velocity w3 and angular momentum L3
ax = plt.figure().add_subplot(projection='3d')
ax.quiver(0, 0, 0, L3[0,0], L3[1,0], L3[2,0], color = 'steelblue', length = 0.1, label = 'L3')
ax.quiver(0, 0, 0, w3[0,0], w3[1,0], w3[2,0], color = 'brown', length = 0.1, label = 'w3')
ax.legend()
```
