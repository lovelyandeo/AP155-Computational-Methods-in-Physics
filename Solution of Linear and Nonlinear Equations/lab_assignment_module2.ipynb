**Calculate the inverse of a matrix and LU decomposition**

1. Use numpy's $\tt linalg$ package to calculate the inverse of the ff. matrix *(5 pts.)*: 

$ \begin{equation} 
{\bf A} = 
\begin{pmatrix} 
4 & -2 & 1 \\
3 & 6 & -4 \\
2 & 1 & 8 
\end{pmatrix} \end{equation} $

2. Check your inverse in both directions; that is, check that ${\bf AA}^{−1} = {\bf A}^{−1}{\bf A} = {\bf I}$. Up to how many number of decimal places is this true? *(5 pts.)*

3. Let this matrix $\bf A$ be used to describe three simultaneous linear equations, ${\bf Ax = b}$. Solve for the vector $\bf x$ given
${\bf b} = \begin{pmatrix} 
12 \\ 
-25 \\ 
32 
\end{pmatrix}$ using: (i) the numerical inverse calculated in (1) and (ii) LU decomposition and backsubstitution with numpy's $\tt linalg$ package. *(5 pts. each)*


**Solution**

**1.1** Since we are using numpy's $\tt linalg$ package, we first have to input numpy itself. Then take the given 3 x 3 matrix, make an array, and compute for the inverse.

```python
# Importing required package
import numpy as np

# Storing the given matrix in an array
A = np.array([[4, -2, 1],
              [3, 6, -4],
              [2, 1, 8]], float)

# Calculating the inverse yung the package
A_inv = np.linalg.inv(A)
print(A_inv)
```

**1.2** To check if ${\bf AA}^{−1} = {\bf A}^{−1}{\bf A} = {\bf I}$, we use numpy's $\tt matmul$ method.

```python
# For AA^-1
print(np.matmul(A,A_inv), "\n")

# For A^-1A
print(np.matmul(A_inv, A))
```

An identity matrix is where the principal diagonal are ones and the rest of the elements are zeros. From the results, we can just approximate the other values as zero. Based on the values, this is true up to the 161t, 17th, and 18th places.

**1.3** Next is we use matrix $A$ to describe three simultaneous linear equations. Solving for $x$ numerically, where in $Ax=b$, we can just multiply the inverse of A on both sides. That will be as follows:

```python
# Numerical inverse calculation
# Ax = b 
# A^-1 Ax = A^-1 b
# Ix = A^-1 b
b = np.array([[12],
              [-25],
              [32]], float)
x = np.matmul(A_inv, b)
print(x)
```

To check the values of $x$, let's now try the other method: the LU decomposition and backsubstition.

The matrices and vectors are already stored as arrays, hence we can move on to the elimination/decomposition right away. It it where the each row of the matrix is normalized, then a backsubstitution is applied to get the exact values of $x$. Or we can just use the $\tt linalg$ function directly from numpy.

```python
from numpy.linalg import solve
x = solve(A,b)
print(x)
```

**Solve for eigenvalues and eigenvectors of a symmetric matrix**

1. Use numpy's $\tt linalg$ package to find the eigenvalues and eigenvectors of the ff. symmetrix matrix *(10 pts.)*: 

$ \begin{equation} 
{\bf A} = 
\begin{pmatrix} 
8 & -2 & -20 \\
-2 & -3 & -9 \\
-20 & -9 & -3 
\end{pmatrix} \end{equation} $

2. Check your answer numerically; that is, check that $\bf AV = VD$, where the $N$ x $N$ matrix $\bf V$ has the eigenvectors as its columns and $\bf D$ is the diagonal matrix with the eigenvalues $\lambda_i$ as its diagonal entries. *(5 pts.)*

**Solution**

**2.1** First is to import the packages needed from $\tt numpy$. And in order to solve for the matrix, we have store its values into a three-dimensionsal array. Then, use the $\tt eigh()$ method to calculate for the eigenvalues and eigenvectors of the matrix.

```python
# Importing necessary libraries
from numpy import array
from numpy.linalg import eigh 

# Storing the matrix in an array
A = array([[ 8, -2, -20 ], 
            [ -2, -3, -9 ], 
            [ -20, -9, -3]], float) 

# Calling the eigh() method to calculate both the eigenvalues and eigenvectors
x,V = eigh(A)

print('Eigenvalues: ', x, "\n") # Printing the eigenvalues
print('Eigenvectors: ', V) # Printing the eigenvectors
```

The code results to two arrays: the one-dimensional array containing the eigenvalues which are $-21.99606195$, $0.16021522$, and $23.83584674$ , and the three-dimensional array whose columns are the corresponding eigenvectors.

**2.2** To check our answer numerically, we show that $\bf AV = VD$. $V$ and $A$ are already defined, hence we just need to construct the diagonal matrix D.

```python
# Left-hand side A V
print('AV: ', np.matmul(A,V), "\n")

# Constructing the diagonal matrix
D = array([[ -21.99606195, 0, 0 ], 
            [ 0, 0.16021522, 0 ], 
            [ 0, 0, 23.83584674]], float) 

# Right-hand side V D
print('VD: ', np.matmul(V,D))
```

As shown, the values from the left-hand side and right-hand side which means that $\bf AV = VD$.

**Solve a nonlinear equation using the relaxation method**

Consider the equation $x=1- e^{-cx}$, where $c$ is a known parameter and $x$ is unknown. This equation arises in a variety of situations, including the physics of contact processes, mathematical models of epidemics, and the theory of random graphs.

1. Write a program to solve this equation for $x$ using the relaxation method for the case $c = 2$. Calculate your solution to an accuracy of at least $10^{-6}$. *(10 pts.)*

2. Modify your program to calculate the solution for values of $c$ from 0 to 3 in steps of 0.01. *(10 pts.)* 

3. Make a plot of $x$ as a function of $c$. Describe the different regimes that you see. *(5 pts.)*

**Solution**

**4.1** Using the relaxation method, we define the given value of $c$ and take an initial guess for $x$. We then perform atleast 1000 iterations to see where the equation converges, and stores the values. Once we compare the old and new value and our desired accuracy has already reached, then we break the loop.

```python
# Importing the necessary function
from math import exp 

# Defining the function
def f(x):
    return 1 - exp(-c*x)

# Defining the given c
c = 2

# Initial guess for x
x = 1

values = []
# Performing the iteration
for i in range(1000):
    values.append(x)
    x0 = x
    x = f(x)

    # Breaking out of the loop 
    while abs(x - values[-1]) > 10**-6: 
        print(f"x={x} after {i} iterations")
        break
```

From the results, we can see that the equation converges to $ x = 0.7968133630966887$ when $c=2$.

**4.2** Next is to get the solution for the values fo $c$ from $0$ to $3$. Modifying the previous code, we first define a function where c is an input, and define the error formula. Then we create an array for the values of $c$ and perform the iteration over that.

```python
# Defining the equation with parameter c
def f(x,c):
    return 1 - exp(-c*x)

# Defining the formula of the error
def error(x,x1,c):
    return abs(f(x1,c)-f(x,c))

# Making an array of c
c = np.linspace(0,3,301)

values = []
# Performing the iteration
for i in c:
    x = 1 # Initial guess
    x1 = f(x,i)
    err = error(x,x1,i)

    while err >=10**-6:
        x = x1
        x1 = f(x,i)
        err = error(x,x1,i)
    values.append([x,i,err])
```

**4.3** Plotting $x$ as a function of $c$, we first have to extract the necessary values from the list. Then customize the plot as follows:

```python
# Importing the necessary libraries for plotting
import matplotlib.pyplot as plt

# Extratcing the c and x values from the data
c = np.array([values[k][1] for k in range(1,len(values))])
x = np.array([values[k][0] for k in range(1,len(values))])

# Plotting the values
fig, ax = plt.subplots()
ax.plot(c,x)

# Plotting the given error for comparison
plt.axvline(1, color='gray', linewidth=3, linestyle = 'dashed') 

# Setting the labels of axes
ax.set_xlabel('c')
ax.set_ylabel('x')

# Displaying the plot
plt.show()
```

As we can see, the values of $x$ are zero up to $c = 1$ then from there onwards, it starts to exponentially increase, showing a clear transition from a regime in which $x = 0$ to a regime of nonzero $x$. Which is also called as percolation transition or epidemic threshold, in physics.

**Physics Application: Calculating the Lagrange point**

_Refer to Exercise 6.16 of the Newman text._ The $L_1$ Lagrange point is the point between the Earth and the Moon at which a satellite will orbit the Earth in perfect synchrony with the Moon, staying always in between the two. 

Assuming circular orbits, and assuming that the Earth is much more massive than either the Moon or the satellite, the distance $r$ from the center of the Earth to the $L_1$ point satisfies
$$ \begin{equation}
\frac{GM}{r^2} - \frac{Gm}{(R-r)^2} = \omega^2 r,
\end{equation} $$
where $R$ is the distance from the Earth to the Moon, $M$ and $m$ are the Earth and Moon masses, $G$ is Newton's gravitational constant, and $\omega$ is the angular velocity of both the Moon and the satellite.

Use either one of (i) Newton's method or (ii) the secant method to solve for the distance r from the Earth to the $L_1$ point. Compute a solution accurate to at least four significant figures. (30 pts.) 

**Solution**

Given in the book, we first define the values of the parameters that will be used.

```python
# Defining the parameters

G = 6.674e-11 # N*m^2/kg^2
M = 5.974e24 # kg
m = 7.348e22 # kg

R = 3.844e8 # m
omega = 2.662e-6 # s^-1
```

Since for this problem, it is possible to solve for the derivative of the equation for the orbit motion given above, then we can use Newton's method which an estimate for the root of a function $x$, uses its slope at a point to then get a better estimate $x'$. The equation of the slope is $f'(x) = \frac{f(x)}{\Delta x}$ and the new estimate for the root is given by
$$x' = x-\Delta x = x-\dfrac{f(x)}{f'(x)}$$


For the given equation of the circular orbit, we first define the equation $f(r)$ as
$$ f(r) = \frac{GM}{r^2} - \frac{Gm}{(R-r)^2} - \omega^2 r$$
and the get its derivative which is
$$f'(r) = -\frac{2GM}{r^3} + \frac{2Gm}{(R-r)^3} - \omega^2 $$
Then, we have to take an initial guess and approximate the function by its tangent line. Next, take the intersection which will become the new $x$ and loop it until it converges to a certain value.

```python
# Defining the equation for the circular orbit
def f(r):
    return (G*M/r**2) - (G*m/(R-r)**2) - (omega**2)*r

# Defining the derivative of the circular orbit equation
def deriv_f(r):
    return (-2*G*M/r**3) - (2*G*m/(R-r)**3) - (omega**2)

# Taking initial guess
r = R/4
# Setting up preferred accuracy
a = 1e-4

# Maximum iteration
max_i = 1000
# Calculating for r using Newton's Method noting the eq for the new estimate
for i in range(max_i+1):
    new_r = -f(r)/deriv_f(r)
    r = r + new_r
    if abs(new_r) <= a:
        break
        
# Printing the output
if i >= max_i:
    print('Maximum iterations has reached')
else:
    print(f'Earth to L1 distance = {r}m')
    
```

Therefore, the distance from the Earth to the L1 point is approximately $326045071\ m$ or approximately $326045\ km$.
